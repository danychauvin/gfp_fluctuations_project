- Distributions of length (birth and division), and fluorescence (#GFP), and growth-rate stratified per GL

2) Variables are extracted in order to denoise the data:
- Growth rate: mean, and variance
- Production rate: mean and variance
- Assymetry at division

3) Data are denoised using input variables

### Prepare YAML file for Moma batch run

````python
import shutil
import os
#Generate a YAML file from experiment_summary_table
#Select subset of interest
subset=experiment_summary_table[experiment_summary_table["promoter"]=="hi1"]
#Batch folder that should be used or created to supervise tracking/curation/export
batch_folder="20190515_preproc_right_20220916_moma_batch"

#Create the batch folder
if not os.path.exists("%s/%s"%(experimentFolderPath,batch_folder)):
    os.mkdir("%s/%s"%(experimentFolderPath,batch_folder))
    print("Making batch directory %s."%(batch_folder))
else:
    print("Batch folder already exist.")
    
if not os.path.exists("%s/%s/mm.properties"%(experimentFolderPath,batch_folder)):
    shutil.copy("./mm.properties","%s/%s"%(experimentFolderPath,batch_folder))
    print("Copying standard mm.properties file to %s."%(batch_folder))
else:
    print("mm.properties already exists, copying cancelled.")
````

````python
#Creating the related yaml file
#For each position, regular expressions are used to find the set of growth lanes that should be added to the YAML file.
preproc_dir="%s/20190515_preproc_right_bis"%(experimentFolderPath)
batch_file="moma_batch_hi1.yaml"
if os.path.exists("%s/%s/%s"%(experimentFolderPath,batch_folder,batch_file)):
    os.remove("%s/%s/%s"%(experimentFolderPath,batch_folder,batch_file))
    print("Batch YAML file already existed and was replaced.")

print("Batch file is created for the following positions, for all detected growth lanes:")


f=open("%s/%s/%s"%(experimentFolderPath,batch_folder,batch_file),"w")
f.write("file_format: 0.1.0\n")
f.write("preprocessing_path: \'%s\'\n"%(preproc_dir))
f.write("default_moma_arg:\n")
f.write("  p: \'%s/%s/mm.properties\'\n"%(experimentFolderPath,batch_folder))
f.write("  analysis: \'%s\'\n"%(batch_file[:-5]))
f.write("pos:\n")

for pos in list(subset["pos"]):
    print("Pos%i"%(pos))
    f.write("  %i:\n    gl:\n"%(pos))
    for root, dirs, files in pyMoM.walklevel("%s/%s"%(preproc_dir,"Pos%i"%(pos)),0):
        gls=sorted([int(e[6+len(str(pos)):]) for e in dirs])
    for i in gls:
        f.write("      %i: {}\n"%(i))
f.close()
````
